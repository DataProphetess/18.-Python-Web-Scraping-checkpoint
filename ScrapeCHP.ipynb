```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://aziza.tn/fr/home"

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup)
```

```python
names, prices, sizes = [], [], []
```

```python
products = soup.findAll("li", class_="product-item")

for product in products[4:]:
    product_name = product.find("a", class_="product-item-link").text.strip()
    product_price = product.find("span", class_="price").text.strip()
    product_size = product.find("span", class_="more_title").text.strip()
    names.append(product_name)
    prices.append(product_price)
    sizes.append(product_size)
```

```python
df = pd.DataFrame({"product_names": names, "product_prices": prices, "product_sizes": sizes})
```
